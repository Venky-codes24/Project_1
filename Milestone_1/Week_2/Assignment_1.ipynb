## Practice on Pandas and Data Visualization

- 1. What is the overall survival rate of passengers aboard the Titanic?

```python
import pandas as pd
import seaborn as sns

# Load Titanic dataset (comes with seaborn)
df = sns.load_dataset("titanic")

# Overall survival rate
total_passengers = len(df)
survivors = df["survived"].sum()
survival_rate = (survivors / total_passengers) * 100

print("Total passengers:", total_passengers)
print("Total survivors:", survivors)
print("Overall survival rate: {:.2f}%".format(survival_rate))
```

- 2. How does the survival rate vary by passenger class (Pclass)?

```python
import seaborn as sns

# Load Titanic dataset
df = sns.load_dataset("titanic")

# Group by passenger class (pclass) and calculate mean survival rate
survival_by_class = df.groupby("pclass")["survived"].mean() * 100
print(survival_by_class)
```

- 3. What is the average age of passengers who survived vs. those who did not survive?

```python
import seaborn as sns

# Load dataset
df = sns.load_dataset("titanic")

# Drop rows with missing ages
df = df.dropna(subset=["age"])

# Calculate average age for survivors and non-survivors
avg_age = df.groupby("survived")["age"].mean()
print(avg_age)
```

- 4. How many passengers embarked from each port (Embarked), and what are their survival rates?

```python
import seaborn as sns
import pandas as pd

# Load Titanic dataset
df = sns.load_dataset("titanic")

# Count passengers by embarkation port
embarked_counts = df["embarked"].value_counts()
print("Passengers by embarkation port:\n", embarked_counts)

# Survival rates by embarkation port
embarked_survival = df.groupby("embarked")["survived"].mean() * 100
print("\nSurvival rate by embarkation port:\n", embarked_survival)
```

- 5. What is the distribution of fare prices paid by passengers, and does fare impact survival chances?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset("titanic")

# ----------------------------
# 1. Distribution of Fares
# ----------------------------
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df["fare"], bins=40, kde=True)
plt.title("Distribution of Passenger Fares (All)")
plt.xlabel("Fare")
plt.ylabel("Count")

plt.subplot(1,2,2)
sns.histplot(df[df["fare"] < 200]["fare"], bins=40, kde=True)
plt.title("Distribution of Passenger Fares (Fare < 200)")
plt.xlabel("Fare")
plt.ylabel("Count")

plt.tight_layout()
plt.show()

# ----------------------------
# 2. Fare vs Survival
# ----------------------------
plt.figure(figsize=(6,5))
sns.boxplot(x="survived", y="fare", data=df)
plt.title("Fare Distribution: Survived vs Did Not Survive")
plt.xlabel("Survival (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show()

# ----------------------------
# 3. Average Fare of Survivors vs Non-Survivors
# ----------------------------
avg_fare = df.groupby("survived")["fare"].mean()
print("Average Fare by Survival Status:")
print(avg_fare)

# ----------------------------
# 4. Scatter Plot: Fare vs Age colored by Survival
# ----------------------------
plt.figure(figsize=(7,6))
sns.scatterplot(x="age", y="fare", hue="survived", data=df, alpha=0.7)
plt.title("Fare vs Age (Color = Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

- 8. Show the age distribution of passengers using a histogram or KDE plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset("titanic")

# Drop rows with missing age values
df = df.dropna(subset=["age"])

plt.figure(figsize=(12,5))

# Histogram
plt.subplot(1,2,1)
sns.histplot(df["age"], bins=30, kde=False, color="skyblue")
plt.title("Age Distribution of Passengers (Histogram)")
plt.xlabel("Age")
plt.ylabel("Count")

# KDE Plot
plt.subplot(1,2,2)
sns.kdeplot(df["age"], fill=True, color="orange")
plt.title("Age Distribution of Passengers (KDE Plot)")
plt.xlabel("Age")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

- 10. Visualize survival count by embarkation port (Embarked) with a stacked bar chart.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from Seaborn
df = sns.load_dataset("titanic")

# Group by 'embarked' and 'survived' to get the count
survival_by_embarked = df.groupby(['embarked', 'survived']).size().unstack(fill_value=0)

# Plot the stacked bar chart
survival_by_embarked.plot(kind='bar', stacked=True, figsize=(8,6), color=['red', 'green'])

plt.title('Survival Count by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Number of Passengers')
plt.legend(['Did not survive', 'Survived'], title='Survival')
plt.xticks(rotation=0)
plt.show()
```

- 11. Display a heatmap of the correlation matrix of numerical features in the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset from Seaborn
df = sns.load_dataset("titanic")

# Select numerical features only
numerical_features = df[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']]

# Compute the correlation matrix
corr_matrix = numerical_features.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
